## Hypothesis 2

### The number of deaths per 100 patients is higher in Vienna compared to Lower Austria among all age groups and all genders.



Figure 2: Total number of deaths per 1000 inhabitants in Lower Austria and Vienna until 10.11.2022 by age group and gender (source of data: https://covid19-dashboard.ages.at/).

### Description

For our first complex hypothesis we used a logarithmic scaling to illustrate the decrease over time. However, that decrease is only clearly visible for the number of daily deaths and the number of occupied intensive beds by Covid-19 patients. Hence we contradict the hyptothesis for the number of occupied normal beds by Covid-19 patients.

We slightly changed the second hypothesis: instead of "number of deaths and hospitalized" (defined in the first submission) we worked with number of deaths only because our dataset is missing information about the gender and age of hospitalized patients. On figure 2 and the interactive visualization (presented in the video "hypothesis_2_video.mov"), the total number of deaths due to COVID-19 per 100 inhabitants was showed for Lower Austria and Vienna per age group and gender. The visualizations confirm the hypothesis that the number of deaths is higher in Vienna compared among all age groups and all genders.
